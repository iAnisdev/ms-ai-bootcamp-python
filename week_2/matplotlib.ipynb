```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5., 0.2)

plt.plot(x, x**4, 'red', x, x*90, 'blue', x, x**3, 'green')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0, 5., 0.2)
x2 = np.arange(0, 5., 0.1)

plt.figure(1)
plt.subplot(211)

plt.plot(x1, x1**4, 'red', x1, x1*90, 'blue', x1, x1**3, 'green' , linewidth = 2.0)

plt.subplot(212)

plt.plot(x2 , np.cos(2*np.pi*x2), 'k')

plt.show()
```
